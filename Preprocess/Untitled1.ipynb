```python
from sklearn.datasets import load_boston
boston = load_boston()
dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfY = pd.DataFrame(boston.target, columns=['MEDV'])
```

```python
dfX["logDIS"] = -np.log(dfX.DIS)
dfX["logNOX"] = np.log(dfX.NOX)
plt.scatter(dfX['logDIS'], dfX['logNOX'])
plt.xlim(-3,0)
plt.ylim(0,-1)
plt.show()
```

```python
DIS = np.array([dfX['logDIS'], dfX['logNOX']])
DIS = DIS.T
cov = np.cov(DIS.T)
print(cov)
w, V = np.linalg.eig(cov)
print(V)
DISidx = DIS.dot(V)
plt.scatter(DISidx[:,0], DISidx[:,1])
plt.xlim(-3,0)
plt.ylim(0.5,-.5)
plt.show()
```

```python
X = np.array([
    [-1, -1],
    [-2, -1],
    [-3, -2],
    [1, 1],
    [2, 1],
    [3, 2]
])
plt.scatter(X[:,0], X[:,1], s=100)
plt.ylim(-3,3)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
Z = pca.transform(X)
Z
w,V = np.linalg.eig(pca.get_covariance())
V.T.dot(X.T).T
```

```python
plt.scatter(Z[:,0], Z[:,1], c = 'r', s=100)
plt.ylim(-3,3)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, whiten=0)
pca.fit(X)
Z = pca.transform(X)
plt.scatter(Z[:,0], Z[:,1], c = 'r', s=100)
w,V = np.linalg.eig(pca.get_covariance())
V.T.dot(X.T).T
plt.ylim(-4,4)
plt.xlim(-4,4)
plt.show()
```
